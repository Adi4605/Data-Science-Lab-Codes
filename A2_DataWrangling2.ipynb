```python
import pandas as pd
import numpy as np
```

```python
data ={
    'Name':['A','B','C','D','E'],
    'Maths':[85,90,np.nan,400,75],
    'Science':[88,92,85,95,np.nan],
    'English':[78,85,80,82,79]
}
df=pd.DataFrame(data)
print("Original Dataset:\n",df)
```

```python
df.isnull().sum()
```

```python
# Outlier Detection
Q1=df['Maths'].quantile(0.25)
Q3=df['Maths'].quantile(0.75)

IQR=Q3-Q1

lower=Q1-1.5*IQR
upper=Q3+1.5*IQR

median=df['Maths'].median()
df['Maths']=np.where(
    (df['Maths']>upper) | (df['Maths']<lower),
    median,
    df['Maths']
)
print("\After Removing Outliers:\n",df)
```

```python
df['Maths'].fillna(df['Maths'].mean(),inplace=True)
df['Science'].fillna(df['Science'].mean(),inplace=True)
print("\nAfter Handling Missing Values:\n",df)
```

```python
df['Maths_Normalized']=(
    (df['Maths']-df['Maths'].min())/
    (df['Maths'].max()-df['Maths'].min())
)
print("\nAfter Normalization:\n",df)
```
