```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text="""Natural Language Processing is a branch of Artificial Intelligence."""
print("Original Text:\n",text)
```

```python
# Tokenization
tokens = word_tokenize(text)

print("\nTokens:\n",tokens)
```

```python
# POS Tagging
pos=pos_tag(tokens)
print("\nPOS Tags:\n",pos)
```

```python
# Stopwords Removal
stop_words=stopwords.words('english')
filtered=[]
for word in tokens:
    if word.lower() not in stop_words:
        filtered.append(word)
print("\nAfter Stopwords Removal:\n",filtered)
```

```python
# Stemming
ps=PorterStemmer()
stemmed=[]
for word in filtered:
    stemmed.append(ps.stem(word))
print("\nStemmed Words:\n")
print(stemmed)    
```

```python
#Lemmatization
lm=WordNetLemmatizer()
lemmatized=[]
for word in filtered:
    lemmatized.append(lm.lemmatize(word))
print("\nLemmatized Words:\n")
print(lemmatized)
```

```python
# TF-IDF
documents=["Natural language processing is important", 
           "Artificial intelligence uses natural language"]
tfidf=TfidfVectorizer()
result=tfidf.fit_transform(documents)
print("\nTF-IDF Matrix:\n")
print(result.toarray())
print("\nFeature Names:\n")
print(tfidf.get_feature_names_out())
```
